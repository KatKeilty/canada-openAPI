```python
#Use as Template for Exploration
#Setup
import sys
sys.path.append('..')

from src.api_client import OpenCanadaAPI
from src.utils import extract_resources_info, search_by_keywords
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plotting
%matplotlib inline
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Initialize API
api = OpenCanadaAPI()
print("✓ Setup complete!")
```

```python
# Cell 2: Search for datasets
# Example: Find employment data
results = api.package_search(query="employment statistics", rows=20)

print(f"Found {results['result']['count']} datasets")
print("\nTop 5 datasets:")
for i, pkg in enumerate(results['result']['results'][:5], 1):
    title = pkg.get('title_translated', {}).get('en', pkg['name'])
    print(f"{i}. {title}")
```

```python
# Cell 3: Explore a specific dataset
# Replace with your chosen dataset ID
dataset_id = "2d90548d-50ef-4802-91f8-c59c5cf68251"
dataset = api.package_show(dataset_id)

# Show metadata
print(f"Title: {dataset['result']['title_translated']['en']}")
print(f"Description: {dataset['result']['notes_translated']['en'][:200]}...")
print(f"Organization: {dataset['result']['organization']['title']}")
print(f"\nNumber of resources: {len(dataset['result']['resources'])}")

# Show available resources
resources_df = extract_resources_info(dataset)
resources_df
```

```python
# Cell 4: Download and analyze a resource
# Pick a CSV or JSON resource from the list above
resource_url = resources_df.iloc[0]['url']  # First resource
resource_format = resources_df.iloc[0]['format']

print(f"Downloading {resource_format} file...")

if resource_format.lower() == 'csv':
    # Download and load CSV
    api.download_resource(resource_url, '../data/temp_data.csv')
    df = pd.read_csv('../data/temp_data.csv')
    
    print(f"\n✓ Loaded {len(df)} rows")
    print("\nFirst few rows:")
    display(df.head())
    
    print("\nColumn info:")
    print(df.info())
    
    print("\nBasic statistics:")
    display(df.describe())
```

```python
# Cell 5: Create visualizations
# Example: If your data has numeric columns
if len(df.select_dtypes(include=['number']).columns) > 0:
    # Distribution of first numeric column
    numeric_col = df.select_dtypes(include=['number']).columns[0]
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    df[numeric_col].hist(bins=30)
    plt.title(f'Distribution of {numeric_col}')
    plt.xlabel(numeric_col)
    plt.ylabel('Frequency')
    
    plt.subplot(1, 2, 2)
    df[numeric_col].plot(kind='box')
    plt.title(f'Box Plot of {numeric_col}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Cell 6: Save processed data
# Save cleaned/processed data for later use
df.to_csv('../data/processed_data.csv', index=False)
print("✓ Processed data saved to data/processed_data.csv")
```
