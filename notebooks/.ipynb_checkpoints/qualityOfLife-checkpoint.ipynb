```python
import sys
sys.path.append('..')

from src.api_client import OpenCanadaAPI
from src.utils import extract_resources_info, search_by_keywords
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plotting
%matplotlib inline
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Initialize API
api = OpenCanadaAPI()
print("✓ Setup complete!")
```

```python
results = api.package_search(query="Quality of Life", rows=10)

print(f"Found {results['result']['count']} datasets")
print("\nTop 10 datasets:")
for i, pkg in enumerate(results['result']['results'][:10], 1):
    title = pkg.get('title_translated', {}).get('en', pkg['name'])
    uuid = pkg.get('id', 'N/A')
    print(f"{i}. {title} (ID: {uuid})")
```

```python
dataset_id = "ed66eb30-8893-11e0-a30d-6cf049291510"
dataset = api.package_show(dataset_id)

# Show metadata
print(f"Title: {dataset['result']['title_translated']['en']}")
print(f"Description: {dataset['result']['notes_translated']['en'][:200]}...")
print(f"Organization: {dataset['result']['organization']['title']}")
print(f"\nNumber of resources: {len(dataset['result']['resources'])}")

# Show available resources
resources_df = extract_resources_info(dataset)
resources_df
```
